```python
import tensorflow as tf
```

# step 1 loading data

```python
mnist=tf.keras.datasets.fashion_mnist
```

```python
(training_images,training_labels),(test_images,test_labels)=mnist.load_data()
```

```python
#normalize wala kam yahan hura
training_images=training_images/255.0
test_images=test_images/255.0
```

# step 2 network architecture

```python
model=tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                  tf.keras.layers.Dense(512,activation=tf.nn.relu),
                                  tf.keras.layers.Dense(10,activation=tf.nn.softmax)
                                 ])
#flattern 2 ya 3 dimension ko 1 dimension array banadega
#relu wali layer mein jo bhi flattern k output hongy wh usky input ban jayengy
#relu agar 0 sy bari value hu tw wh value return krta ha agar 0 ya usse choti hu tw 0 return krta ha
#softmax all numbers ka sum kryn and har number ki probablity return krta hajiski probablity ziada hogi wh class hogi
```

# step 3 network compile

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
#Adam same as rmsprop  
#sparse_categorical_crossentropy qu k multi classification problem ha
```

```python
model.fit(training_images,training_labels,epochs=5,batch_size=128)
#train_images and train_labels sy learning hogi
#epochs mein iterations kitny iteration mein learning hogi
#batch size ek sth kitni images jayegi
model.evaluate(test_images,test_labels)
#evaluate mein test hota unseen data yahan py performance and accuracy wala scene ha 
```
